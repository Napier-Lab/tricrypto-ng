```python
import pandas as pd

# plotting:
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
data = pd.read_csv("../../data/marginal_ir.csv")
data['time_elapsed'] = [1.0]*len(data['time_to_maturity']) - data['time_to_maturity']
data['ir_idx'] = data['ir_idx'] * 100
for i in range(1,4):
    data['ir_pt' + str(i)] = data['ir_pt' + str(i)] * 100
data
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 6))

ax.plot(
    data.time_elapsed,
    data.ir_idx,
    'b.',
    label="ir_idx"
)

ax.plot(
    data.time_elapsed,
    data.ir_pt1,
    'r.',
    label="ir_pt1"
)
ax.plot(
    data.time_elapsed,
    data.ir_pt2,
    'g.',
    label="ir_pt2"
)
ax.plot(
    data.time_elapsed,
    data.ir_pt3,
    'y.',
    label="ir_pt3"
)

ax.set_xlabel("time elapsed", fontsize=20)
ax.set_ylabel("Marginal Interest Rate [%]", fontsize=20)
ax.tick_params(axis='both', which='major', labelsize=20)
ax.tick_params(axis='x', which='major', colors='black')
# ax.set_ybound(-10, 100)
ax.legend(fontsize=20)
```

```python
fig.savefig("../../data/ir_analysis.png", dpi=300)
```
